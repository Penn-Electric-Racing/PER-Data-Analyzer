```python
import perda
```

```python
csvPath = "csv_files/data_logs04thMay05-28.csv"

aly = perda.create()
aly.reset()
```

```python
aly.read_csv(csvPath, input_align_name="default")
```

```python
packCurrent = ["ams.pack.current"]

aly.plot(packCurrent, start_time=0, end_time=-1, time_unit="s")
```

```python
# Align variables for timestamp match

brakePressure = [
    "ludwig.brakePressureRaw.front",
    "ludwig.brakePressureRaw.rear",
]

aly.plot(brakePressure, start_time=0, end_time=-1, time_unit="s")

aly.plot(["ludwig.brakePressureRaw.front"], start_time=0, end_time=-1, time_unit="s")

aly.plot(["ludwig.brakePressureRaw.rear"], start_time=0, end_time=-1, time_unit="s")
```

```python
brakePressureRawFront = aly._analyzer__csvparser.get_np_array(
    "ludwig.brakePressureRaw.front"
)
brakePressureRawRear = aly._analyzer__csvparser.get_np_array(
    "ludwig.brakePressureRaw.rear"
)


brakePressurePSIFront = (brakePressureRawFront[:, 1].copy() - 0.5) * 1000 / (4.5 - 0.5)
brakePressurePSIRear = (brakePressureRawRear[:, 1].copy() - 0.5) * 1000 / (4.5 - 0.5)

brakePressureRawFront[:, 1] = brakePressurePSIFront
brakePressureRawRear[:, 1] = brakePressurePSIRear

aly.plot(
    [brakePressureRawFront, brakePressureRawRear],
    start_time=0,
    end_time=-1,
    time_unit="s",
)

aly.plot([brakePressureRawFront], start_time=0, end_time=-1, time_unit="s")

aly.plot([brakePressureRawRear], start_time=0, end_time=-1, time_unit="s")
```
